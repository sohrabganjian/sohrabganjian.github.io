# Perspectives on Quantum Computing: Selected Algorithms and the Evolving Quantum Computing Landscape

The study of quantum mechanics began in the early 1900s with attempts to understand how nature behaves at atomic and subatomic scales and to explain phenomena that classical physics could not, including black-body radiation, the photoelectric effect, and atomic structure. Over time, as quantum mechanics became better understood, scientists began to explore whether its principles could also be used to process and protect information. One of the earliest proposed applications in this direction was Stephen Wiesner's quantum money, developed around 1970 but not [published until 1983](https://dl.acm.org/doi/10.1145/1008908.1008920). The proposal used the fact that arbitrary unknown quantum states cannot be copied perfectly and that measuring a state in an incompatible basis can disturb it. If the issuing bank knows the correct measurement bases while a counterfeiter does not, these properties can be used to construct a cryptographic scheme intended to prevent counterfeiting. Around the same time, scientists also noticed fundamental limitations of classical computation for simulating quantum phenomena. As the relationship between quantum mechanics and information processing became better understood, scientists investigated whether controllable quantum systems could simulate atomic, molecular, and other quantum systems more efficiently. This formed an important early motivation behind developing quantum computers. Over time, researchers expanded their focus to investigate whether these systems could accelerate or provide benefits to areas beyond the simulation of quantum phenomena. This line of work was further driven forward by the inception of Shor's algorithm, which showed that a quantum computer could theoretically factor integers much more efficiently than the best known classical algorithms, even though integer factorization is not inherently a quantum-physics problem. Since then, quantum computing has expanded considerably, and quantum algorithms have been developed for tasks ranging from quantum chemistry and physics to unstructured search, optimization, and linear algebra. 

In this post, we begin by reviewing selected quantum algorithms that use quantum phenomena such as superposition, interference, and entanglement with the aim of solving certain problems more efficiently than classical methods. Although these algorithms represent only a small subset of quantum algorithms, they cover several foundational ideas that also appear in other quantum algorithms. For each algorithm reviewed, a corresponding companion post provides further technical discussion and implementation details. A more extensive collection of quantum algorithms is maintained by the [Quantum Algorithm Zoo](https://quantumalgorithmzoo.org/). We also connect the reviewed algorithms to other work and isolate common techniques, including superposition, interference, entanglement, phase kickback, and amplitude amplification, to help identify patterns across quantum algorithms.

The discussion then shifts from algorithms to the current quantum computing landscape. We examine the quantum software ecosystem and how software helps translate theoretical ideas into programs that can be simulated, compiled, and executed on quantum hardware. We then consider the present state of quantum hardware, technology roadmaps, commercial cloud access, and current pricing models, including concrete prices that help place the cost of quantum computation into perspective. This discussion also considers what the current landscape means for researchers and raises questions about privacy, security, data sovereignty, and dependence on cloud infrastructure.

The discussion also includes a note for those working within quantum science and those engaging with the field from outside it. For researchers, it emphasizes the importance of communicating results according to the available theoretical and experimental evidence, since overstating the capabilities of quantum computing may weaken trust in the field. For those outside the field, it recognizes that quantum mechanics is counterintuitive and encourages open-mindedness, curiosity, and asking questions. Finally, we connect the material to the larger picture by considering what quantum computing has achieved, the problems currently being studied, and possible directions for the near and long-term future.

The aim is to provide a high-level perspective rather than an exhaustive technical treatment, allowing readers to gain a grounded perspective on what quantum computation can and cannot currently do, how the field is evolving, and which questions may be useful when evaluating its progress.

### Oracle Quantum Algorithms

An oracle algorithm is an algorithm that solves a problem by interacting with a black-box subroutine, called an *oracle*, whose internal implementation is abstracted away. Each query to the oracle provides information about how the function evaluates certain inputs, and the goal is to infer some feature of interest, such as whether the function has a global property, whether it contains hidden structure, or which inputs satisfy a specified condition. An oracle algorithm operates within the query model, which is a model of computation where the input is accessed through calls to an oracle. The cost of the algorithm is measured by the number of times the algorithm queries the oracle. In the quantum query model, the oracle is often represented by a unitary $U_f$ that encodes the function, and it acts as
$$
U_f |x\rangle |y\rangle = |x\rangle |y \oplus f(x)\rangle .
$$

As described in the [IBM Quantum query algorithms course](https://quantum.cloud.ibm.com/learning/en/courses/fundamentals-of-quantum-algorithms/quantum-query-algorithms/introduction), 

> The query model of computation is like a petri dish for quantum algorithmic ideas. It's rigid and unnatural in the sense that it doesn't accurately represent the sorts of computational problems we generally care about in practice, but it has nevertheless proved to be incredibly useful as a tool for developing quantum algorithmic techniques. This includes the ones that power the most well-known quantum algorithms, such as Shor's algorithm for integer factorization. The query model also happens to be a very useful framework for explaining quantum algorithmic techniques.

We will briefly review several instances of oracle quantum algorithms: Deutsch's algorithm, Deutsch–Jozsa, Bernstein–Vazirani, Simon's problem, and Grover's search algorithm.
 
### Deutsch's algorithm

A simple example of an oracle quantum algorithm is Deutsch's algorithm, [published in 1985](https://royalsocietypublishing.org/rspa/article-abstract/400/1818/97/15743/Quantum-theory-the-Church-Turing-principle-and-the), which determines whether a Boolean function $f: \{0,1\} \rightarrow \{0,1\}$ is constant or balanced. Deutsch's algorithm can deterministically determine whether the input function $f$ is constant or balanced using one query, whereas a deterministic classical algorithm requires two queries.

### Deutsch–Jozsa Algorithm

A few years after Deutsch's algorithm, Deutsch and Jozsa developed an extension of it in [1992](https://royalsocietypublishing.org/rspa/article-abstract/439/1907/553/16819/Rapid-solution-of-problems-by-quantum-computation), now known as the Deutsch–Jozsa algorithm. This algorithm extends the problem from a single-bit Boolean function to an $n$-bit Boolean function $f_n:\{0,1\}^n \rightarrow \{0,1\}$, while still asking whether the function is constant or balanced. Deutsch–Jozsa requires only one quantum query. In the deterministic worst case, this gives an exponential separation from classical query complexity, since a classical algorithm may need $2^{n-1}+1$ queries to certify that the function is constant. Even in the best classical case, identifying a balanced function with certainty requires observing both output values, so at least two queries are needed. It is also worth noting that, for large $n$, the number of balanced functions is larger than the number of constant functions, which makes average-case discussions more subtle.

### Bernstein–Vazirani

In [1993](https://dl.acm.org/doi/10.1145/167088.167097), the Bernstein–Vazirani algorithm addressed a different type of problem about Boolean functions, namely determining a hidden string $s$. The oracle encodes a function

$$
f_s: \{0,1\}^n \rightarrow \{0,1\},
$$

which is promised to satisfy $f_s(x) = x \cdot s \bmod 2$, where $s$ is a hidden $n$-bit string and $\cdot$ denotes the bitwise inner product. The Bernstein–Vazirani algorithm can recover $s$ with one quantum query, whereas a classical algorithm requires $\mathcal{O}(n)$ queries, giving a linear query speedup. The Bernstein–Vazirani circuit is shown below.

<figure style="text-align: center;">
  <img src="Circuit%20Visualizations/Bernstein_Vazirani_Oracle_Circuit_with_title.jpg" width="700">
  <figcaption><b>Figure 4.</b> Bernstein–Vazirani algorithm circuit.</figcaption>
</figure>

An interesting observation is that the Deutsch–Jozsa and Bernstein–Vazirani circuits have the same structure. The difference lies in what each oracle encodes. The same Hadamard–oracle–Hadamard pattern can answer different questions about Boolean functions, leading to different query-complexity separations. This raises a broader question about how much of a quantum algorithm's power comes from the circuit template itself, and how much comes from the structure encoded in the oracle.

### Simon's Problem

Building on earlier oracle algorithms, Simon's problem, [published in 1994](https://ieeexplore.ieee.org/document/365701), posed a different problem in the oracle model. Here the function $f: \{0,1\}^n \rightarrow \{0,1\}^m$, with $m \geq n$, is implemented as an oracle. The function is promised to be either one-to-one, or there exists a nonzero secret string $s \in \{0,1\}^n$ such that, for all distinct $x,x' \in \{0,1\}^n$, $f(x)=f(x')$ if and only if $x \oplus s = x'$. In the second case, each output value is shared by exactly two inputs separated by the hidden shift $s$, so the function is two-to-one. The problem is to determine which case holds and, in the two-to-one case, recover the hidden string $s$. The figure below shows the circuit for Simon's problem.

<figure style="text-align: center;">
  <img src="Circuit%20Visualizations/Simon_Oracle_Circuit_with_title.jpg" width="700">
  <figcaption><b>Figure 5.</b> Simon's circuit for equal-size input and output registers.</figcaption>
</figure>

Simon's algorithm uses $\mathcal{O}(n)$ oracle queries, followed by classical post-processing to solve a system of linear equations over $\mathbb{F}_2$ and recover the string $s$. Using standard Gaussian elimination, this step is upper bounded by $\mathcal{O}(n^3)$, so the overall procedure remains polynomial-time. In comparison, classical randomized algorithms require on the order of $2^{n/2}$ queries in the oracle model, so Simon's algorithm gives an exponential query speedup.

Simon's algorithm served as an important precursor to Shor's factoring algorithm because it showed how quantum computation could extract hidden periodic structure from an oracle. This hidden-period perspective was later used in Shor's algorithm. Indeed, there was a possibility that Simon's and Shor's work would be merged into a single paper, as noted by Simon in his recollection of events [here](https://web.archive.org/web/20260413200009/https://aws.amazon.com/blogs/quantum-computing/simons-algorithm/). For more information about the development of Shor's algorithm, see [Appendix 2 of this post](https://sohrabganjian.github.io/blog/2026/shors_algorithm/).

### Grover's Algorithm 

Grover's search algorithm, [published in 1996](https://dl.acm.org/doi/10.1145/237814.237866), is an oracle quantum algorithm that shifted attention from problems of mainly theoretical relevance toward tasks with practical utility. Alongside Shor's algorithm, it helped make clear that quantum algorithms were not only useful for theoretical separations, but could also point toward computational problems of practical importance. These early results played an important role in increasing interest in quantum computation and shaping how quantum computing was viewed. 

Grover's algorithm addresses the unstructured search problem, where one is given access to a Boolean function $f:\{0,1\}^n \rightarrow \{0,1\}$ and wants to find an input $x$ such that $f(x)=1$. In this setting, $f$ acts as a marking oracle for candidate solutions. In the formulation considered here, there is a single marked input satisfying $f(x)=1$, but Grover's algorithm can also be extended to cases with multiple marked inputs. Grover's search circuit is depicted below.


<figure style="text-align: center;">
  <img src="Circuit%20Visualizations/GroverCircuit_with_title.jpg" width="700">
  <figcaption><b>Figure 6.</b> Grover's search circuit.</figcaption>
</figure>

Grover's search algorithm gives a quadratic query speedup for unstructured search, reducing the number of oracle queries from $\mathcal{O}(N)$ in the classical query model to $\mathcal{O}(\sqrt{N})$ in the quantum query model. In the presence of additional structure, such as sorted order, classical algorithms can make use of that structure, and the quantum advantage from Grover's algorithm may vanish.

While Grover's algorithm can be viewed as an oracle algorithm, it was also later studied as part of a larger family of algorithms known as amplitude amplification algorithms, further studied by Brassard et al. in [*Quantum Amplitude Amplification and Estimation*](https://arxiv.org/abs/quant-ph/0005055). Abstractly, as is also the case in Grover's algorithm, amplitude amplification algorithms are built around an operator of the form $\mathcal{Q} = \mathcal{A}\mathcal{S}_0\mathcal{A}^\dagger\mathcal{S}_f$, where $\mathcal{A}$ prepares the initial state, $\mathcal{S}_0$ flips the phase of the all-zero state, and $\mathcal{S}_f$ flips the phase of basis states for which a chosen predicate $f(x)$ equals $1$. More concretely, suppose $\mathcal{A}$ prepares $\mathcal{A}|0\rangle = |\psi\rangle$. The state $|\psi\rangle$ can be decomposed into a "good" component, corresponding to inputs satisfying the predicate, and a "bad" component, corresponding to the remaining inputs,

$$
|\psi\rangle = \sqrt{p}|\psi_{\text{good}}\rangle + \sqrt{1-p}|\psi_{\text{bad}}\rangle .
$$

Here $p$ is the initial success probability. The goal of amplitude amplification is to increase the probability of measuring a state from the good component. This is done by applying $\mathcal{Q}$ a near-optimal number of times. Applying it too few times gives limited amplification, while applying it too many times can rotate the state past the good component and reduce the success probability. In Grover's algorithm, $\mathcal{A}=H^{\otimes n}$, so the reflection $\mathcal{A}\mathcal{S}_0\mathcal{A}^\dagger$ becomes the Grover diffuser $H^{\otimes n}S_0H^{\otimes n}$. For a search space of size $N$ with a single marked input, the Grover iterate is applied on the order of $\sqrt{N}$ times to obtain a high probability of measuring the marked input. Note that the query complexity of Grover's algorithm is also $\mathcal{O}(\sqrt{N})$. [Quantum counting](https://en.wikipedia.org/wiki/Quantum_counting_algorithm) is another algorithm that uses principles from amplitude amplification.

To summarize, oracle quantum algorithms played an important role in the early development of quantum computing. The techniques developed in these algorithms later influenced algorithms such as Shor's algorithm, while Grover's search also became a useful algorithm for search problems. For a further survey on quantum oracle algorithms, see [*Oracle Quantum Algorithms: An Overview*](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=4862127).

### Fourier and Phase-Based Quantum Algorithms

Quantum computers can encode useful information in the relative phases of basis states, while suitable unitary transformations can introduce phases associated with their eigenstates. Through interference, these phases can be converted into changes in the amplitudes of quantum states, making target states more likely to be observed upon measurement. The quantum Fourier transform (QFT) and quantum phase estimation (QPE) provide systematic ways of extracting this kind of phase and periodic structure. They appear as fundamental tools in many quantum algorithms, including Shor's algorithm and the Harrow–Hassidim–Lloyd (HHL) algorithm, which we briefly review next.

### Quantum Fourier Transform 

The quantum Fourier transform maps computational basis states to Fourier basis states. As a change of basis, QFT is most useful when it is applied to states with phase or periodic structure, since it can convert that structure into amplitude patterns that become accessible through measurement. For this reason, QFT is usually treated as a subroutine or algorithmic primitive rather than as a standalone problem-solving algorithm. It appears, for example, in quantum phase estimation and Shor's algorithm.

QFT can be expressed mathematically as a unitary acting on a quantum state in an $N=2^n$ dimensional Hilbert space,

$$
\mathrm{QFT}_N:\mathbb{C}^N \rightarrow \mathbb{C}^N,
$$

defined on computational basis states by

$$
\mathrm{QFT}_N|x\rangle =
\frac{1}{\sqrt{N}}
\sum_{k=0}^{N-1} e^{\frac{2\pi i xk}{N}}|k\rangle,
\qquad x \in \{0,\ldots,N-1\}.
$$

For each input basis state $|x\rangle$, the phase appearing in each Fourier component $|k\rangle$ depends on both the input value $x$ and the Fourier basis value $k$. When QFT is used as a subroutine, these relative phase differences can cause constructive interference at frequencies compatible with the hidden period and destructive interference elsewhere, making periodic information more accessible.

### Quantum Phase Estimation 

In many settings, knowing the eigenvalue corresponding to an eigenstate of a unitary $U$ provides useful information. QPE aims to solve this problem. Given a unitary $U$ and an eigenstate $|u\rangle$, where

$$
U|u\rangle = \lambda |u\rangle,
$$

the goal is to estimate the eigenvalue $\lambda$. Since $U$ is unitary, its eigenvalues have the form $\lambda = e^{2\pi i\theta}$ for some $\theta \in [0,1)$. QPE estimates the phase $\theta$, and the target precision $\epsilon$ affects the number of qubits and controlled applications of $U$ required. This is why it is called phase estimation.

At a high level, QPE works by preparing a phase register in superposition, followed by the application of controlled powers of $U$, namely $\mathrm{Cont}\text{-}U$, $\mathrm{Cont}\text{-}U^2$, $\mathrm{Cont}\text{-}U^4$, $\ldots$, $\mathrm{Cont}\text{-}U^{2^{t-1}}$, where $t$ is the number of phase register qubits. Because the target state is an eigenstate of $U$, each controlled application introduces a phase depending on $\theta$, and these phases accumulate across the phase register. The inverse QFT is then applied to turn this phase pattern into a computational basis state encoding an approximation of $\theta$. Finally, the phase register is measured to obtain an estimate of $\theta$. In compact form, QPE performs the following operation:

$$
|0\rangle^{\otimes t}|u\rangle \xrightarrow{\mathrm{QPE}} |\widetilde{\theta}\rangle |u\rangle.
$$

It uses controlled powers of $U$ to imprint the eigenphase onto an auxiliary register, then uses the inverse QFT to make that phase information measurable.

QPE appears in several important quantum algorithms, most notably Shor's algorithm. It also appears in HHL, quantum chemistry algorithms for estimating ground-state energies, quantum counting, and Hamiltonian simulation, among other examples. The quantum phase estimation circuit can be seen below.

### HHL Algorithm

The Harrow–Hassidim–Lloyd (HHL) algorithm, introduced by Aram Harrow, Avinatan Hassidim, and Seth Lloyd, with a [journal publication in 2009](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.103.150502), is a quantum algorithm for solving linear systems of equations. Given a system $A\boldsymbol{x}=\boldsymbol{b}$, HHL aims to prepare a quantum state proportional to the solution vector, $|x\rangle \propto A^{-1}|b\rangle$. If one wants to reconstruct the full classical solution vector using state tomography, the quantum advantage may vanish. The speedup is instead relevant when one wants to estimate global properties of the solution state, or when HHL is used as a subroutine inside a larger quantum algorithm. For the speedup to hold, several input assumptions are needed. For example, $A$ is assumed to be sparse, to have a low condition number, and to allow efficient implementation of the unitary $e^{iAt}$ through Hamiltonian simulation. For well-formed inputs satisfying these assumptions, and for tasks that only require selected global or statistical information about the solution state rather than full classical reconstruction, HHL provides an exponential speedup over the best known classical approaches. These assumptions limit the practical applicability of HHL. For more discussion of considerations such as input preparation, output readout, sparsity, condition number, and whether the desired information can be extracted without losing the speedup, see Scott Aaronson's [*Quantum Machine Learning Algorithms: Read the Fine Print*](https://www.scottaaronson.com/papers/qml.pdf).

Regardless of these assumptions, the techniques used in HHL are valuable because they show how quantum computation can be applied to linear algebra and spectral methods. The algorithm uses the spectral decomposition $A = \sum_{i=0}^{N-1} \lambda_i |u_i\rangle \langle u_i|$, where $\{|u_i\rangle\}_{i=0}^{N-1}$ is an eigenbasis of $A$. The input state $|b\rangle$ can be written in the same eigenbasis as $|b\rangle = \sum_{i=0}^{N-1} b_i |u_i\rangle$. Therefore, solving $A|x\rangle = |b\rangle$ for $|x\rangle$ gives

$$
\begin{aligned}
| x \rangle &= \frac{A^{-1} |b\rangle}{\| A^{-1} |b \rangle \|_2} \\
&= \frac{1}{ \| A^{-1} |b \rangle \|_2 } \sum\limits_{i=0}^{N-1} \frac{1}{\lambda_i} | u_i \rangle \langle u_i | \sum\limits_{j=0}^{N-1} b_j | u_j \rangle  \\
&= \frac{1}{\sqrt{\sum\limits_{i=0}^{N-1} \frac{|b_i|^2}{\lambda_i^2}}} \sum\limits_{i = 0}^{N - 1} \frac{b_i}{\lambda_i} |u_i \rangle. 
\end{aligned}
$$

At a high level, HHL prepares the input state $|b\rangle$, uses Hamiltonian simulation and QPE to access eigenvalue information about $A$, performs a controlled operation corresponding to eigenvalue inversion, and then uncomputes the phase estimation register. After postselection, the resulting state is proportional to $|x\rangle \propto A^{-1}|b\rangle$ as a quantum state.

This algorithm shows the potential of quantum computers for linear algebra and spectral analysis, since HHL decomposes $|b\rangle$ into eigencomponents of $A$ and applies a transformation depending on each eigenvalue, producing a state proportional to $\sum_i \frac{b_i}{\lambda_i}|u_i\rangle$. Related ideas were later developed in frameworks such as quantum singular value transformation, which can implement many other linear algebraic transformations, in a more general setting.

### Shor's Algorithm

Shor's algorithm, first introduced [by Peter Shor in 1994](https://ieeexplore.ieee.org/document/365700), is one of the most groundbreaking quantum algorithms. It gives a polynomial-time quantum algorithm for integer factorization, with major consequences for cryptography. In particular, cryptosystems such as RSA rely on the assumption that factoring large integers is computationally infeasible for classical computers. A scalable, fault-tolerant quantum computer running Shor's algorithm would render such cryptosystems insecure. Although such a quantum computer has not yet been built, Shor's algorithm showed that quantum computation could theoretically make this possible. In response to this long-term threat, the U.S. National Institute of Standards and Technology (NIST) [launched its post-quantum cryptography standardization project in 2016](https://web.archive.org/web/20260507233444/https://nvlpubs.nist.gov/nistpubs/ir/2016/NIST.IR.8105.pdf). Post-quantum cryptography aims to develop cryptographic schemes based on problems that are expected, based on current evidence, to remain difficult for both classical and quantum computers.

Shor's algorithm consists of classical pre-processing and post-processing, while the main computational work is handled by the quantum order-finding subroutine. Given an integer $N$ to be factored, one chooses an integer $x \in \mathbb{Z}_N^*$, meaning $x$ is coprime to $N$, and studies the periodic function $f(k)=x^k \bmod N$. The quantum subroutine uses modular exponentiation together with QFT or phase estimation to extract information about the period $r$, also called the order of $x$ modulo $N$. Contingent on satisfying additional conditions, this period can then be used classically to find nontrivial factors of $N$ by computing $\gcd(x^{r/2}-1,N)$ and $\gcd(x^{r/2}+1,N)$. Shor's algorithm is probabilistic, so a single run may fail, but repeated runs can give a high probability of success.

### Variational Quantum Algorithms 

Many quantum algorithms use a fixed circuit template with predefined gates. Certain gates may depend on the input instance, as in HHL or Shor's algorithm, but once those choices are made, the circuit implements a fixed unitary for that instance. Variational quantum algorithms are different in that they use a parameterized family of circuits $U(\boldsymbol{\theta})$, called an *ansatz*. Here, the ansatz refers to the chosen circuit structure, while a specific parameter value $\boldsymbol{\theta}$ determines a particular unitary within that family. For each parameter value, the circuit is run many times to estimate a cost function, and an optimizer then updates $\boldsymbol{\theta}$ for the next iteration. At a high level, a variational quantum algorithm prepares a parameterized quantum state $|\psi(\boldsymbol{\theta})\rangle = U(\boldsymbol{\theta})|0\rangle$, where $U(\boldsymbol{\theta}) \in \mathbb{C}^{2^n \times 2^n}$ is an $n$-qubit unitary and $\boldsymbol{\theta} \in \mathbb{R}^d$, with $\boldsymbol{\theta}=(\theta_1,\theta_2,\ldots,\theta_d)$, is a vector of trainable parameters. The parameter dimension $d$ is determined by the chosen ansatz and does not need to match the Hilbert space dimension $2^n$. Depending on the circuit design, $d$ may be smaller than, equal to, or larger than $2^n$.

The goal of a variational quantum algorithm is to adjust the parameters of the ansatz so that a problem-specific cost function is optimized. This cost function is usually written as the expectation value of an observable $\hat{O}$,

$$
C(\boldsymbol{\theta})=\langle \psi(\boldsymbol{\theta})|\hat{O}|\psi(\boldsymbol{\theta})\rangle.
$$

Here $\hat{O}$ denotes the observable whose expectation value defines the quantity being optimized. A typical variational workflow alternates between quantum circuit execution and classical optimization. The quantum computer prepares $|\psi(\boldsymbol{\theta}^{(k)})\rangle$ many times to estimate the cost value $C(\boldsymbol{\theta}^{(k)})$ from measurement samples, while the optimizer updates the parameters,

$$
|\psi(\boldsymbol{\theta}^{(1)})\rangle \rightarrow C(\boldsymbol{\theta}^{(1)}) \rightarrow
|\psi(\boldsymbol{\theta}^{(2)})\rangle \rightarrow C(\boldsymbol{\theta}^{(2)}) \rightarrow
\cdots \rightarrow
|\psi(\boldsymbol{\theta}^{(T)})\rangle \rightarrow C(\boldsymbol{\theta}^{(T)}).
$$

After several iterations, the optimized parameter vector $\boldsymbol{\theta}^{(T)}$ defines a circuit intended to prepare a state that gives a near-optimal solution to the problem at hand.

Variational quantum algorithms are widely used in quantum machine learning, although quantum machine learning also includes non variational approaches such as quantum kernel methods, quantum support vector machines, and quantum linear algebra-based methods. Whether variational quantum algorithms and quantum machine learning provide broad advantage over classical algorithms remains an open research question. Challenges include noise, measurement cost, barren plateaus, ansatz design, optimizer behaviour, competitive and effective classical heuristics, and limited theoretical guarantees.

There is also an important structural difference between quantum machine learning models and classical neural networks. In classical machine learning, nonlinear activation functions such as sigmoid, tanh, and ReLU are inserted between layers, and the architecture can freely change dimension from one layer to the next. In contrast, the coherent part of a quantum circuit is unitary. It is linear, reversible, and maps a $2^n$ dimensional Hilbert space to itself. As a result, a variational quantum circuit does not have a direct analogue of intermediate nonlinear activation layers. In quantum machine learning, nonlinear behaviour can enter through data encoding, measurement statistics, cost functions, postselection, or classical post-processing, but this is different from placing a controlled nonlinear activation function after each layer of a classical neural network. This makes direct comparisons between quantum machine learning models and classical neural networks subtle. It also suggests that useful quantum machine learning may not come from simply copying classical neural network architectures, but from using operations that are natural for quantum computers. For more discussion, see PennyLane's posts [*Why measuring performance is our biggest blind spot in quantum machine learning*](https://pennylane.ai/blog/2022/03/why-measuring-performance-is-our-biggest-blind-spot-in-quantum-machine-learning/) and [*What could quantum computing actually be useful for in machine learning?*](https://pennylane.ai/blog/2026/03/quantum-computing-useful-for-machine-learning).

Variational quantum algorithms are especially natural for problems that are quantum in nature, meaning problems where the object being studied is itself a quantum system. Examples include quantum chemistry, condensed matter physics, and Hamiltonian simulation. They are also viewed as promising near-term approaches for noisy intermediate scale quantum (NISQ) devices for several reasons, including their ability to use relatively shallow circuits and to work within hybrid quantum-classical workflows. Although the added value of variational quantum algorithms relative to existing classical methods remains an active research question, there are still several reasons for continued interest in variational quantum algorithms. For example, existing research suggests that some quantum models can have high expressive power compared with classical counterparts, as discussed in [*The power of quantum neural networks*](https://arxiv.org/pdf/2011.00027). For a broader review of variational quantum algorithms, see [*Variational quantum algorithms*](https://www.nature.com/articles/s42254-021-00348-9) by Cerezo et al.

We will briefly review two major early variational quantum algorithms, the variational quantum eigensolver (VQE) and the quantum approximate optimization algorithm (QAOA), both of which helped establish the variational quantum algorithm landscape.

### Variational Quantum Eigensolver

The variational quantum eigensolver (VQE) is one of the first major variational quantum algorithms for a problem of direct interest in quantum chemistry and quantum physics, namely approximating the ground-state energy of a Hamiltonian operator. The first demonstration of VQE by Peruzzo et al. in [*A variational eigenvalue solver on a quantum processor*](https://arxiv.org/abs/1304.3061) involved estimating the ground-state energy of the helium hydride cation $\mathrm{HeH}^+$. The VQE problem can be stated as follows:

> Given a Hamiltonian $H$ describing a quantum system, find an approximation of its ground-state energy $E_0$ and a corresponding parameterized quantum state $|\psi(\boldsymbol{\theta})\rangle$ such that
>
> $$E_0 \approx \min_{\boldsymbol{\theta}} \langle \psi(\boldsymbol{\theta}) | H | \psi(\boldsymbol{\theta}) \rangle,$$
>
> where $\boldsymbol{\theta}$ denotes the variational parameters.

The reason the expectation value does not go below the ground-state energy $E_0$ is due to the Rayleigh–Ritz variational principle, which states that for any normalized trial state $|\psi\rangle$, $\langle \psi | H | \psi \rangle \geq E_0$. This motivates the use of parameter optimization to drive the energy estimate toward $E_0$. Estimates of this kind have applications in materials science, molecular binding energy calculations, reaction energetics, and molecular properties such as polarizability, among others.

The VQE workflow begins with a Hamiltonian, which is the energy operator describing the quantum system and its dynamics. For molecular systems, this Hamiltonian includes the behaviour of both electrons and nuclei. To make the problem computationally manageable, one usually applies approximations that simplify the interaction between electronic and nuclear motion, with the Born–Oppenheimer approximation being a common choice. The resulting electronic Hamiltonian is then expressed in a finite basis and mapped from a fermionic Hamiltonian to a qubit Hamiltonian, represented as a sum of Pauli operators. This allows the problem to be encoded on a gate-based quantum computer using a parameterized ansatz. The choice of ansatz and initial parameters can make a substantial difference. Some choices may make the optimization landscape harder to navigate, while others may guide the algorithm more easily toward low-energy states. The algorithm then uses repeated circuit evaluations and an optimizer to move through the energy landscape toward a low-energy solution.

For a more detailed review of VQE methods and techniques, see [*The Variational Quantum Eigensolver: a review of methods and best practices*](https://arxiv.org/abs/2111.05176). The VQE workflow is shown below.

### Quantum Approximate Optimization Algorithmhm 

Shortly after the introduction of VQE, Edward Farhi et al. introduced the [quantum approximate optimization algorithm](https://arxiv.org/abs/1411.4028), which extended the scope of variational quantum algorithms to optimization problems. QAOA provided a variational framework for studying combinatorial optimization problems on gate-based quantum computers. QAOA is inspired by adiabatic quantum computation, where one begins with an easy-to-prepare ground state and slowly evolves toward a Hamiltonian whose low-energy states encode near-optimal solutions to an optimization problem. QAOA adapts this idea to gate-based quantum computers by replacing the continuous evolution with alternating parameterized operations. In a two-term setting, the time-dependent Hamiltonian interpolates between two Hamiltonian components as
$$
H(t)=A(t)H_M+B(t)H_C,\quad t\in[0,T],
$$
where $H_C$ is the cost Hamiltonian encoding the optimization problem, $H_M$ is the mixer Hamiltonian, $T$ is the total evolution time, and $A(t)$ and $B(t)$ are smooth functions satisfying $A(0)=1$, $B(0)=0$, $A(T)=0$, and $B(T)=1$.

The mixer Hamiltonian is chosen so that its ground state is easy to prepare, while the cost Hamiltonian is constructed so that low-energy states correspond to near-optimal solutions of the optimization problem. In an adiabatic process, the system begins in the ground state of the mixer Hamiltonian and slowly evolves toward the cost Hamiltonian. If the evolution is slow enough and some additional conditions are satisfied, the final state remains close to the ground state of the cost Hamiltonian. On a gate-based quantum computer, the time evolution generated by the continuous Hamiltonian $H(t)$ must be approximated using a sequence of unitary gates. A common approximation method is Trotterization, where evolution under a sum of Hamiltonian terms is replaced by alternating unitary evolutions generated by the individual terms. This leads to the QAOA ansatz

$$
U(\boldsymbol{\gamma},\boldsymbol{\beta}) =
\prod_{k=1}^{p} e^{-i\gamma_k H_C} e^{-i\beta_k H_M},
$$

where $\boldsymbol{\gamma}=(\gamma_1,\ldots,\gamma_p)$ and $\boldsymbol{\beta}=(\beta_1,\ldots,\beta_p)$ are trainable parameters. The number of layers $p$ is chosen as a hyperparameter. Smaller values of $p$ may not be expressive enough, while larger values can increase circuit depth and may make optimization more difficult. In the limit $p \rightarrow \infty$, QAOA may recover the corresponding continuous adiabatic evolution under an appropriate choice of parameters.

### Transformation Frameworks 

While many quantum algorithms perform specific linear algebraic tasks, such as using HHL to apply an inverse eigenvalue transformation in a linear system, there are also more general frameworks for implementing families of such transformations. We can view quantum signal processing (QSP) and quantum singular value transformation (QSVT) as transformation frameworks. They provide systematic ways to implement polynomial transformations of quantities encoded in quantum operations, subject to certain validity conditions on the polynomial and the encoding.

These ideas were first developed in the setting of QSP, where one implements polynomial transformations of a scalar signal parameter encoded in a unitary operation, and later extended to QSVT, where polynomial transformations are applied to the singular values of a block-encoded matrix. Schematically, the big picture can be expressed as

$$
\mathrm{QSP}: x \mapsto p(x), 
\qquad
\mathrm{QSVT}: A \mapsto p^{(\mathrm{SV})}(A),
$$

where $x$ is a scalar signal parameter, $A$ is an encoded operator, and $p^{(\mathrm{SV})}(A)$ denotes applying $p$ to the singular values of the encoded operator $A$.

### Quantum Signal Processing

QSP allows scalar polynomial transformations to be embedded inside a unitary operation. In a QSP formulation with $d+1$ phase angles, the target polynomial $p(x)$ has degree at most $d$, satisfies $|p(x)| \leq 1$ for all $x \in [-1,1]$, and has parity matching $d$, meaning it contains either only even or only odd powers. Functions or polynomials that do not satisfy these conditions can be handled by polynomial approximation, normalization, restricting the domain, or decomposing the target into even and odd parts and combining the resulting constructions. Therefore, QSP can be used for a wider range of target functions than only polynomials that satisfy these conditions.

Using an alternating sequence of the signal operator

$$
U(a)=\begin{bmatrix}
a & \sqrt{1-a^2}\\
\sqrt{1-a^2} & -a
\end{bmatrix},
$$

and the signal-processing operator

$$
S(\phi)=\begin{bmatrix}
e^{i\phi} & 0\\
0 & e^{-i\phi}
\end{bmatrix},
$$

for a fixed $a$ and phases $\phi_0,\ldots,\phi_d$, QSP constructs an operator of the form

$$
S(\phi_0)\prod_{k=1}^d U(a)\,S(\phi_k) =
\begin{bmatrix}
p(a) & g_{01}(a)\\
g_{10}(a) & g_{11}(a)
\end{bmatrix},
$$
where $p(a)$ is the evaluation of $p(x)$ at $x=a$! The entries $g_{01}(a), g_{10}(a), g_{11}(a)$ are determined by expanding the product, but they are not of interest. Thus, the value $p(a)$ is embedded inside a larger unitary operation. Although selected amplitudes can be estimated through measurement procedures, QSP is usually valuable as a building block inside larger algorithms. For example, it can be used within Hamiltonian simulation algorithms. QSP is also the scalar foundation behind QSVT, where the same idea is extended from transforming a scalar parameter $a$ to transforming the singular values of an encoded matrix.

### Quantum Singular Value Transformation 

In QSVT, the goal is to apply a polynomial transformation to the singular values of an encoded matrix $A$. Let the singular value decomposition of $A$ be

$$
A = W \Sigma V^\dagger,
$$

where

$$
\Sigma = \operatorname{diag}(\sigma_1,\sigma_2,\dots,\sigma_r),
\qquad
\sigma_1\geq \sigma_2\geq \cdots \geq \sigma_r \geq 0.
$$

Schematically, for an odd polynomial $p$, QSVT implements transformations of the form

$$
A = W\Sigma V^\dagger
\quad\longmapsto\quad
W p(\Sigma) V^\dagger,
$$

where $p$ is a polynomial chosen to implement the desired singular value transformation. Even polynomials take a slightly different form, although the underlying goal remains the same, namely applying $p$ to the singular values of $A$. For further details, see Section 3.3 of [*Quantum Singular Value Transformation and Beyond*](https://arxiv.org/abs/1806.01838). 

Applying $p$ to the singular values is represented by

$$
p(\Sigma)=\operatorname{diag}(p(\sigma_1),p(\sigma_2),\dots,p(\sigma_r)).
$$

This is why the framework is called quantum singular value transformation: QSVT transforms the singular values of the encoded matrix while keeping the singular-vector basis fixed.

QSVT does not require $A$ to be square. In general, $A$ may be a rectangular matrix $A\in\mathbb{C}^{m\times n}$. 

The matrix $A$ is encoded as a block of a larger unitary $U_A$, called a block-encoding of $A$. This encoding is not unique. One can think of $U_A$ as having the block structure

$$
U_A = \begin{bmatrix} A/\alpha & * \\ * & * \end{bmatrix},
$$

where $\alpha\geq \|A\|$ is a normalization factor, with $\|A\|$ denoting the largest singular value of $A$. The value of $\alpha$ also depends on the encoding.

To see how QSVT works, consider one choice of block-encoding for a normalized matrix $A$ satisfying $\|A\|\leq 1$,

$$
U(A)=
\begin{bmatrix}
A & \sqrt{I-AA^\dagger}\\
\sqrt{I-A^\dagger A} & -A^\dagger
\end{bmatrix}.
$$

The QSVT analogue of the QSP signal-processing operator is a projector-controlled phase operator. In block form, this operator can be written schematically as

$$
\Pi_\phi =
\begin{bmatrix}
e^{i\phi}I & 0\\
0 & e^{-i\phi}I
\end{bmatrix}.
$$

Depending on the parity of the polynomial $p$, that is, whether $p$ is even or odd, the QSVT sequence takes slightly different forms.

For a polynomial of even degree $d$, the QSVT sequence has the form

$$
\left[
\prod_{k=1}^{d/2}
\Pi_{\phi_{2k-1}} U(A)^\dagger \widetilde{\Pi}_{\phi_{2k}} U(A)
\right]
\Pi_{\phi_{d+1}}
=
\begin{pmatrix}
p^{(\mathrm{SV})}(A) & * \\
* & *
\end{pmatrix}.
$$

For a polynomial of odd degree $d$, the QSVT sequence has the form

$$
\widetilde{\Pi}_{\phi_1} U(A)
\left[
\prod_{k=1}^{(d-1)/2}
\Pi_{\phi_{2k}} U(A)^\dagger \widetilde{\Pi}_{\phi_{2k+1}} U(A)
\right]
\Pi_{\phi_{d+1}}
=
\begin{pmatrix}
p^{(\mathrm{SV})}(A) & * \\
* & *
\end{pmatrix}.
$$

Here $U(A)$ is the block-encoding unitary. The operators $\Pi_\phi$ and $\widetilde{\Pi}_\phi$ are projector-controlled phase operators defined by

$$
\Pi_\phi=e^{i\phi}\Pi+e^{-i\phi}(I-\Pi),
\qquad
\widetilde{\Pi}_\phi=e^{i\phi}\widetilde{\Pi}+e^{-i\phi}(I-\widetilde{\Pi}).
$$

Here $\Pi$ and $\widetilde{\Pi}$ are orthogonal projectors onto the subspaces used to encode the input and output spaces of $A$.

Thus, QSVT generalizes QSP by replacing the scalar signal $a$ with a block-encoded matrix $A/\alpha$. In this sense, QSP can be viewed as the special case of QSVT where the encoded matrix is $1\times 1$, namely a scalar $a\in[-1,1]$.

QSVT is not just a technique for matrix inversion. Its strength comes from the flexibility of choosing the polynomial transformation. Many important quantum algorithms can be reinterpreted as polynomial transformations of spectral data, which is why QSVT is often labelled as ["A Grand Unification of Quantum Algorithms"](https://arxiv.org/abs/2105.02859). In this view, Grover search, Hamiltonian simulation, phase estimation, and HHL-like matrix inversion are not isolated tricks, but different choices of block-encoding and polynomial transformation. In short, QSVT gives a general framework for performing transformations on singular values.

The selected quantum algorithms reviewed here represent only a small portion of the quantum algorithms landscape. They are foundational examples, and the techniques used in them appear in many other quantum algorithms. Many of the algorithms reviewed here also have their own extensions, refinements, and adaptations to different applications. For example, Grover adaptive search can be used for optimization problems such as constrained polynomial binary optimization (CPBO), as demonstrated in [*Grover Adaptive Search for Constrained Polynomial Binary Optimization*](https://arxiv.org/abs/1912.04088). Ideas already present in Simon's algorithm helped shape the development of Shor's algorithm, and Shor's algorithm connects naturally to the broader family of hidden subgroup algorithms. These examples show how quantum algorithms are connected to one another. Even when a particular algorithm does not directly target a problem of real-life utility, or cannot yet be implemented at a practical scale because of hardware limitations, it can still introduce techniques and perspectives that reappear in other quantum algorithms with different goals.

Beyond these examples, there is a plethora of other quantum algorithms, subroutines, and applications. ### Common Quantum Algorithmic Techniques

While each quantum algorithm solves a specific problem, there are some common techniques that appear across many of them. We will briefly review some of these commonly used techniques:

**Superposition and interference.** Superposition is one common element among quantum algorithms intended to achieve a meaningful speedup over classical algorithms. Through superposition, states are prepared with amplitudes spread across many basis states. A quantum state can be written as a superposition of computational basis states,

$$
|\psi\rangle=\sum_x \alpha_x |x\rangle,
\qquad
\sum_x |\alpha_x|^2=1,
$$

where $\alpha_x$ is the amplitude corresponding to the basis state $|x\rangle$. It is important to note that although operations can act on a superposition of many basis states, the result of measurement yields only one outcome. As such, the algorithm must shape the amplitudes through interference so that states associated with the desired information become more likely to be observed. If two amplitude contributions $\alpha$ and $\beta$ are associated with the same final basis state, the probability of that basis state is

$$
|\alpha+\beta|^2 = |\alpha|^2 + |\beta|^2 + 2\mathrm{Re}(\alpha^*\beta),
$$

not simply $|\alpha|^2+|\beta|^2$. The extra term $2\mathrm{Re}(\alpha^*\beta)$ is the interference term. Depending on the relative phases of $\alpha$ and $\beta$, this term can be positive, leading to constructive interference, or negative, leading to destructive interference. Quantum superposition and interference are two common ingredients for quantum advantage over classical algorithms.

**Entanglement.** Entanglement describes correlations between quantum systems that cannot be written as independent states of the individual parts. For two registers $A$ and $B$, a product state has the form

$$
|\psi\rangle_{AB}=|\phi\rangle_A\otimes|\chi\rangle_B.
$$

An entangled state cannot be written in this form. For example, the Bell state

$$
|\Phi^+\rangle=\frac{|00\rangle+|11\rangle}{\sqrt{2}}
$$

is entangled because there are no single-qubit states $|\phi\rangle$ and $|\chi\rangle$ such that $|\Phi^+\rangle=|\phi\rangle\otimes|\chi\rangle$. Entanglement creates non-classical correlations between subsystems and appears in many quantum algorithms. For example, in Shor's algorithm, the phase and modular exponentiation registers become entangled during the order-finding subroutine. A 2002 result by Richard Jozsa and Noah Linden, in [*On the role of entanglement in quantum computational speed-up*](https://arxiv.org/abs/quant-ph/0201143), showed that for quantum algorithms operating on pure states to achieve exponential speedup, multipartite entanglement growing with the input size is necessary.

**Phase kickback.** Phase kickback is a core quantum mechanism that allows phase information, such as eigenphase information encoded in a unitary, to be transferred onto a control register. This transforms hidden phase information into relative phases in the control register, which can subsequently be decoded into a classical result. If $U|\psi\rangle=e^{i\phi}|\psi\rangle$, then

$$
\mathrm{Ctrl}\text{-}U\left(\alpha|0\rangle+\beta|1\rangle\right)|\psi\rangle =
\left(\alpha|0\rangle+\beta e^{i\phi}|1\rangle\right)|\psi\rangle.
$$

The phase $e^{i\phi}$ has effectively been kicked back onto the control register. In QPE, the same idea is applied using controlled powers of $U$. If $|u\rangle$ is an eigenstate of $U$ with eigenvalue $e^{2\pi i\theta}$, so that $U|u\rangle=e^{2\pi i\theta}|u\rangle$, then these controlled powers create a phase pattern across the phase register. The inverse QFT then converts this phase pattern into a computational basis estimate of $\theta$. In compact form, QPE performs

$$
|0\rangle^{\otimes t}|u\rangle \xrightarrow{\mathrm{QPE}} |\widetilde{\theta}\rangle |u\rangle.
$$


Phase kickback is a fundamental technique used in many important quantum algorithms, particularly those involving eigenvalue estimation or phase estimation, such as QPE, Shor's algorithm, and the HHL algorithm.


**Amplitude amplification.** Amplitude amplification increases the probability of measuring states satisfying a chosen predicate. If

$$
|\psi\rangle=\sqrt{p}|\psi_{\mathrm{good}}\rangle+\sqrt{1-p}|\psi_{\mathrm{bad}}\rangle,
$$

then the amplification operator $\mathcal{Q}=\mathcal{A}\mathcal{S}_0\mathcal{A}^{\dagger}\mathcal{S}_f$ rotates the state toward the good component, increasing the probability of observing a desired outcome. Grover's search is the main example, while similar ideas show up in amplitude estimation and quantum counting.

**Uncomputation.** Quantum algorithms often use auxiliary registers to store intermediate information. In some cases, these auxiliary registers become entangled with the working register, which is the register intended to hold the relevant output of the computation. To ensure the correctness of the procedure, the working register may need to be disentangled from the auxiliary registers before measurement or before later steps of the algorithm.

Suppose the goal is to produce $|x\rangle|f(x)\rangle$, but computing $f(x)$ requires intermediate information stored in an auxiliary register,

$$
|x\rangle|0\rangle|0\rangle
\longmapsto
|x\rangle|g(x)\rangle|f(x)\rangle.
$$

Here $|g(x)\rangle$ represents temporary workspace used during the computation. The uncomputation trick applies the inverse of the computation that produced this auxiliary state, returning the auxiliary register to $|0\rangle$ while preserving the desired output. Schematically,

$$
|x\rangle|0\rangle|0\rangle
\longmapsto
|x\rangle|g(x)\rangle|f(x)\rangle
\longmapsto
|x\rangle|0\rangle|f(x)\rangle.
$$

This technique is used in HHL after the eigenvalue inversion step. The phase estimation, or clock, register is uncomputed so that it no longer remains entangled with the system register that carries the solution state.

### The Quantum Software Ecosystem

As quantum computing grows, quantum algorithms are continually being introduced, refined, optimized, and extended. Quantum software plays an important role in this process by providing an interface between theoretical ideas and hardware execution. It helps translate algorithms into circuits, circuits into hardware-aware implementations, and quantum workloads into executable programs whose results can be analyzed, validated, and used for scientific or practical purposes, eventually beyond what can be obtained through classical simulation. This makes quantum software an important part of the path toward quantum utility. 

Open-source quantum software is especially valuable at this stage of the field, where scalable fault-tolerant quantum computers have not yet been realized. Open-source tools make it easier to study not only the algorithms themselves, but also the layers that connect algorithms to implementation, including circuit optimization, qubit routing, layout selection, error mitigation, error suppression, simulation, and application-specific workflows such as quantum chemistry. This openness supports knowledge exchange among experts, students, early-career researchers, software developers, domain researchers, and curious learners from different backgrounds, helping make quantum computing more accessible while allowing the community to test ideas, compare methods, and build toward applications with practical value.

As quantum computing and quantum software development evolve, the development of the software ecosystem itself can be used as one indicator of the growth and maturity of the field. Software packages do not only provide convenient interfaces for writing quantum circuits. They also signal where the field is moving. As hardware improves and research priorities change, some modules, subroutines, and abstractions may fade away, while others, such as primitives for hardware execution, error mitigation, error correction, compilation, resource estimation, and integration with high-performance computing, receive more attention.

Qiskit provides a useful example of this evolution. IBM released [Qiskit v0.1 in March 2017](https://www.ibm.com/quantum/qiskit/history) as an open-source framework for programming quantum computers. In 2018, IBM released the early [Qiskit elements](https://www.ibm.com/quantum/qiskit/history), including Terra for core circuit building, Aqua for domain-specific use cases, Ignis for early error-mitigation work, and Aer for quantum simulation. These reflected the needs of the time, including circuit construction, simulation, early error mitigation, and application-oriented algorithm modules. As the field matured, Qiskit moved toward a more stable and performance-oriented software stack. The release of [Qiskit 1.0](https://www.ibm.com/quantum/blog/qiskit-1-0-release-summary) marked a shift toward performance, stability, and usability, with IBM noting that Qiskit 1.0 was intended to support circuits with 100+ qubits and lay groundwork for future 1,000+ qubit workloads. The later release of [Qiskit 2.0](https://www.ibm.com/quantum/blog/qiskit-2-0-release-summary) continued this direction by emphasizing performance improvements, removal of deprecated features and legacy interfaces, and closer interaction with high-performance computing communities. These changes suggest a field moving from small demonstrations toward larger workloads, better software engineering, and closer coordination between quantum and classical computing.


The educational and application layers around Qiskit also reflect this direction. IBM Quantum Learning now includes courses on topics such as [quantum algorithms, quantum error correction, quantum machine learning, variational algorithm design, quantum chemistry with VQE, utility-scale quantum computing, and integrating quantum with high-performance computing](https://quantum.cloud.ibm.com/learning/en/courses). This is another sign that quantum software is no longer only about constructing small circuits, but also about building workflows, training users, and connecting algorithms to realistic hardware and application contexts.

PennyLane shows a related but different direction. Developed by Xanadu, PennyLane has focused strongly on differentiable quantum programming, variational algorithms, quantum machine learning, and quantum chemistry. Its ecosystem includes [quantum datasets](https://pennylane.ai/datasets/), [quantum chemistry tools and workflows](https://pennylane.ai/quantum-chemistry), and software for compiling hybrid quantum-classical programs. For example, [Catalyst](https://docs.pennylane.ai/projects/catalyst/en/stable/) enables just-in-time compilation of PennyLane programs and is designed to compile entire quantum-classical workflows. PennyLane's Lightning simulators also point toward high-performance simulation and integration with classical computing resources. Xanadu also develops related open-source tools such as [The Walrus](https://the-walrus.readthedocs.io/en/latest/), which supports photonic quantum computing and Gaussian boson sampling. Together, these tools show how quantum software is developing around hybrid workflows, specialized applications, and quantum programs that account for both hardware constraints and classical computing resources.

This software layer plays an essential role in evaluating quantum utility because it helps us understand what a hardware-aware transpiled circuit would look like, how circuit complexity scales with problem size, and whether the resulting computation may still be classically simulatable using advanced classical methods, including tensor-network techniques, multi-GPU simulation, or parallel computing. The simulability of quantum circuits is important because, as of now, there is no general practical method for deciding whether an arbitrary quantum circuit is classically simulatable. Clifford circuits, for example, are efficiently classically simulatable by the Gottesman–Knill theorem, and even some circuits containing non-Clifford gates may remain simulatable depending on their structure, depth, entanglement, or tensor-network complexity. Quantum software is therefore part of how the field tests whether a proposed quantum computation is genuinely moving beyond classical simulation and helps guide researchers toward problems where quantum hardware may provide meaningful value.

While the multitude of quantum software packages and specialized tools is valuable, it also creates challenges. Researchers and developers need to know which tools already exist, how different packages represent quantum circuits, and whether a workflow built in one framework can be moved to another without being rewritten from scratch. This is important because duplicated effort can slow progress. Some compatibility between tools is possible through plugins, but these connections can be fragile when packages update. In this light, languages such as [OpenQASM](https://openqasm.com/) are valuable. OpenQASM, short for Open Quantum Assembly Language, is a quantum assembly language originally introduced in [2017](https://arxiv.org/abs/1707.03429) for describing quantum circuits and experiments. For example, a circuit written in one framework may be exported to OpenQASM and then imported into another framework, depending on the features supported by each package. OpenQASM therefore helps provide a common language, but it does not remove all compatibility challenges. Still, its development reflects the growing need for shared representations as quantum software moves from isolated packages toward more connected workflows.

### Current Quantum Computing Landscape 

The current quantum computing landscape is shaped by the NISQ (Noisy Intermediate-Scale Quantum) era, a term coined by John Preskill in his 2018 paper, [*Quantum Computing in the NISQ era and beyond*](https://arxiv.org/abs/1801.00862). While quantum computers with modest qubit counts exist at a scale beyond brute-force classical simulation, their overall capabilities and scalability are constrained by hardware limitations such as physical noise and limited coherence times.

That said, current quantum hardware has already been used in meaningful demonstrations of quantum utility, especially in hybrid quantum-classical workflows. For example, in the Science Advances paper [*Chemistry beyond the scale of exact diagonalization on a quantum-centric supercomputer*](https://www.science.org/doi/10.1126/sciadv.adu9991), researchers used a hybrid quantum-classical workflow based on sample-based quantum diagonalization (SQD) to study electronic-structure problems using quantum circuits that are difficult to simulate classically. In the largest case, the Iron-Sulfur Cluster $[4\mathrm{Fe}\text{-}4\mathrm{S}]$ was encoded using 72 qubits for the electronic-structure problem and embedded into a 77-qubit hardware execution with auxiliary qubits. The experiment used circuits with up to 10,570 gates, including 3,590 two-qubit gates, and produced approximate ground-state energy estimates and upper bounds in a regime beyond the reach of brute-force exact diagonalization. Similar examples involving IBM Quantum research can be found on the [IBM Quantum blog](https://www.ibm.com/quantum/blog). Similarly, IonQ partnered with Airbus to study the aircraft cargo loading problem, a constrained variation of the NP-hard knapsack problem. Using a variant of QAOA, they deployed problem instances ranging from 12 to 28 qubits on trapped-ion hardware to identify candidate cargo distributions, obtaining optimal solutions for the instances studied. The studied instances were small enough that the output could be validated against known optimal solutions and simulator results. This experiment serves how an optimization problem with practical relevance can be formulated, run on current trapped-ion hardware, and tested as a candidate workflow for scaling with future hardware. More information can be found in IonQ's post on [quantum cargo loading optimization](https://www.ionq.com/resources/learn-quantum-cargo-loading-optimization) and the corresponding paper [*Quantum Computing for Optimizing Aircraft Loading*](https://arxiv.org/abs/2504.01567).


These demonstrations reflect a shift in the NISQ era. The field is moving away from basic, isolated test circuits toward complex experiments that integrate physical quantum hardware into hybrid workflows alongside high-performance classical computing and methods tailored to specific problems. Whether such demonstrations provide a quantum advantage over the best known classical methods is a more subtle question, but they support the view that quantum computing experiments are moving from small proof-of-principle circuits toward physical implementations with increasing practical relevance. This is evident in efforts to push hardware scale beyond classical simulation, and in experiments showing the practicality and scalability of quantum computation. Developing these workflows and establishing utility on near-term systems serves as a stepping stone toward the realization of fault-tolerant quantum computing.

One challenge of the current NISQ era is finding applications that fit within present hardware constraints. Many of the most powerful theoretical quantum algorithms are based on QFT or QPE, but these methods are difficult to run at useful scale on noisy hardware. QPE requires controlled powers of a unitary, such as $U, U^2, \ldots, U^{2^{t-1}}$, and the required circuit depth grows quickly with the desired precision. Since current hardware is noisy and coherence times are limited, errors accumulate throughout these deep controlled circuits and can reduce the accuracy of the final result. The challenge of executing such algorithms is also reflected in experimental implementations of Shor's algorithm. As noted in IonQ's blueprint report [*Fault-Tolerant Quantum Computing with Trapped Ions: The Walking Cat Architecture*](https://arxiv.org/abs/2604.19481), the largest integer factored using a hardware implementation that performs the required modular arithmetic remains 15, first demonstrated in [2001](https://www.nature.com/articles/414883a). These observations may explain why many current experimental demonstrations instead use variational and sampling-based approaches, hybrid methods, or problem-specific methods.

Even beyond the NISQ era and into the fault-tolerant era, quantum computers will likely rely on hybrid computation, with different parts of a problem executed across quantum processors and classical resources. Classical infrastructure is expected to remain integrated into operations such as system control, compilation, quantum error correction decoding, and processing results. IBM's approach of [quantum-centric supercomputing](https://www.ibm.com/think/topics/quantum-centric-supercomputing) reflects this direction by connecting quantum processors with classical high-performance computing infrastructure and coordinating computation across them.

Current quantum computing providers are pursuing a variety of hardware platforms, including superconducting qubits, trapped ions, neutral atoms, photonic systems, quantum annealers, and topological approaches. This diversity is valuable because different platforms have different strengths and engineering challenges. Some may offer faster gates, others higher connectivity, longer coherence times, easier fabrication, or more natural access to certain problem classes. Since it is not yet clear which hardware approach will scale most effectively to fault-tolerant quantum computing, pursuing multiple platforms gives the field more paths forward.

Many providers also publish roadmaps describing their expected progress in qubit count, gate fidelity, error correction, logical qubits, and fault-tolerant architectures. For example, IBM has announced a roadmap toward fault-tolerant quantum computing by 2029. Interestingly, superconducting qubits, which IBM uses in its processors, face substantial engineering challenges when scaled to larger systems. These include maintaining sufficiently low physical error rates, producing high-quality qubits consistently at scale, integrating control hardware, and managing the physical-qubit overhead of quantum error correction, as explored in [*How to Build a Quantum Supercomputer: Scaling from Hundreds to Millions of Qubits*](https://arxiv.org/abs/2411.10406). Each quantum hardware platform has its own strengths and tradeoffs, which provides a reason for pursuing multiple hardware platforms.

The differences in access methods, pricing structures, and billing units across providers show that using quantum hardware is not yet straightforward or standardized. The differences extend beyond cost and availability because each provider also has its own hardware and software requirements. A logical circuit may be portable across providers when a compatible software framework performs the required compilation for the selected hardware. However, its executable form generally cannot be transferred directly because different providers differ in their native gates, hardware connectivity, submission formats, and supported features. The total turnaround time also extends beyond execution on the quantum processor. It can include circuit compilation, queueing, server-side processing, quantum execution, error mitigation, transfer of results, and additional classical post-processing as part of a hybrid computation. A quantum processor may also become temporarily unavailable because of maintenance or operational interruptions.

Quantum hardware providers generally report current device availability and announce planned maintenance, while some publish recent incidents affecting their cloud platforms. However, detailed statistics for individual processors, such as historical uptime, average queue times, job failure rates, and end-to-end latency, are not consistently made public. This can make it difficult for users to estimate costs and plan research timelines. The study [*Quantum Computing in the Cloud: Analyzing Job and Machine Characteristics*](https://arxiv.org/abs/2203.13121) provides statistics on the operational performance of IBM Quantum systems. The study was published in 2022 and reflects an earlier generation of quantum cloud services, but it still provides useful context for understanding these operational considerations.

The current quantum computing landscape is heavily centred around cloud access, and this trend will likely continue for the foreseeable future. This raises concerns about privacy, data sovereignty, and the protection of sensitive computations. A submitted quantum program may expose its input data, circuit structure, parameters, execution metadata, or results to the service provider. Organizations working with classified or commercially sensitive information may therefore require dedicated or on-premises quantum systems governed by appropriate data storage, access, and security policies. Cryptographic techniques such as blind quantum computation and quantum homomorphic encryption may help address these concerns. In their original forms, these protocols require limited quantum capabilities on the client side and quantum communication with the server. Classical-client variants also exist, but they require additional cryptographic assumptions or protocol arrangements and currently involve substantial computational and communication overhead. Consequently, these techniques are not yet technologically practical at the scale required for real-world quantum computations.

In an increasingly polarized and rapidly changing geopolitical environment, dependence on foreign quantum infrastructure may itself become a national security risk. Even access provided by an allied country could be restricted following political disagreement, changes in export controls, sanctions, or national security decisions. Since quantum expertise and infrastructure cannot be developed on short notice, sustained investment in domestic quantum technologies is a reasonable measure for protecting national interests, maintaining technological independence, and avoiding excessive reliance on countries that may control access to advanced quantum computing.

One possible long-term approach to reducing dependence on quantum cloud services is the development of desktop quantum computers, although building such systems at a computationally useful scale presents an even greater engineering challenge and remains far beyond current capabilities. Nevertheless, SpinQ has already developed a [commercial three-qubit desktop quantum computer](https://arxiv.org/abs/2202.02983) that operates at room temperature. The system is designed primarily for educational purposes and provides direct experience with quantum hardware rather than demonstrating quantum utility. Although this device is not a scalable alternative to cloud quantum computers, it shows that quantum hardware can take a compact form suitable for local operation. Following this direction, one can imagine a personal computer in which CPUs, GPUs, and a local QPU work together through an integrated computing environment, although only time will tell whether such a device will ever be built. However, the history of humanity is full of examples of overcoming challenges once thought impossible, such as stepping on the moon or running a sub-two-hour marathon. These examples are not evidence that a useful personal quantum computer will necessarily be built, but they leave room for scientific ambition and the possibility of overcoming challenges that currently seem out of reach.

### A Note for Those Within and Beyond Quantum Science

Scientists working in quantum technologies and adjacent fields have a responsibility to communicate the capabilities of quantum technologies accurately and set expectations accordingly. In simplified descriptions, assumptions, and conditions may sometimes be omitted, producing claims stronger than the underlying scientific results. The following examples contrast several commonly stated claims with what is supported by the available scientific evidence.

For Shor's algorithm,

> **Drifted statement:** Shor's algorithm solves integer factorization exponentially faster than any classical algorithm.
>
> **Established scientific claim:** Shor's algorithm solves integer factorization *superpolynomially* faster than the best *known* classical algorithms.

The best known classical factoring algorithms have superpolynomial but subexponential complexity. Additionally, no theoretical result rules out a polynomial-time classical factoring algorithm. That said, discovering one would have profound consequences for cryptography and theoretical computer science.

For HHL,

> **Drifted statement:** HHL solves systems of linear equations exponentially faster than classical computers.
>
> **Established scientific claim:** Under specific input and output assumptions, HHL efficiently prepares a quantum state encoding the solution and allows selected properties of that state to be estimated.

Recovering the complete solution as classical data may cause the claimed speedup to vanish.

For Grover's algorithm,

> **Drifted statement:** Grover's algorithm provides a quadratic speedup for search.
>
> **Established scientific claim:** Grover's algorithm provides a quadratic query speedup for unstructured search.

When additional structure is available, such as sorted order, classical algorithms can use that structure, and Grover's quadratic advantage no longer applies.

As a good scientific practice, researchers should state the assumptions and conditions under which a result holds. This helps researchers remain aware of the scope of their own claims and sets expectations according to the available scientific evidence. Simplifications, whether made consciously or unconsciously, can produce claims stronger than the underlying research supports. For established results such as Shor's algorithm, an individual simplification may be relatively harmless. However, repeated simplifications can accumulate and make imprecise language feel acceptable. Their effects on judgement, expectations, and scientific reasoning may not be immediately visible, but they can introduce bias and produce unintended consequences over time. This is not about being pedantic, but about presenting an honest account of the capabilities and limitations of quantum computing. Such precision is particularly important when communicating with early-career researchers or nontechnical audiences, who may otherwise form an impression of quantum computing that differs substantially from its scientific reality. Over time, overstated claims may distort public understanding and weaken trust in the field rather than strengthen it. Scott Aaronson's [*Quantum Machine Learning Algorithms: Read the Fine Print*](https://www.scottaaronson.com/papers/qml.pdf) provides a particularly relevant discussion of this problem.

At the same time, conducting research in quantum science comes with its own challenges. These may include studying counterintuitive phenomena that have little connection to everyday experience, conducting lengthy experiments, or simply identifying interesting research problems. Taking risks is an inevitable part of life and research. It is therefore reasonable to spend time becoming familiar with the field, developing intuition, and formulating meaningful scientific questions. Even when not everything is clear, it is fine to take a risk and attempt to solve the problem without tying its value to whether it demonstrates quantum advantage. Whatever the outcome, it is important to reflect, learn, and grow so you can improve for the next time.

For those interested in quantum computing but not actively working in the field, including people with a non-quantum scientific background, the simply curious, or those who interact with quantum technologies through management, administration, policy, or law, it is completely understandable to find quantum concepts counterintuitive. They describe phenomena that we do not usually experience in everyday life, and depending on one's background, developing intuition for them may take time. Many resources, including some mentioned throughout this post, can help build intuition and provide useful visualizations. Furthermore, maintain perspective. Please keep in mind that quantum computing and quantum technologies are not going to solve all of the world's problems. They are powerful tools with huge implications, but expectations should remain grounded in scientific evidence. Let empirical evidence guide your decisions and viewpoints, rather than an assumption that anything labeled 'quantum' is inherently better. 

Just be curious, ask questions and try to enjoy the process!

In this post, we covered considerable ground, including a review of selected quantum algorithms, an overview of other algorithms in the field, and several techniques that appear across quantum computation. Beyond the algorithms themselves, we discussed the current quantum software ecosystem and its role in developing and testing algorithms, assessing whether quantum circuits remain classically simulatable, and identifying problems that may provide utility beyond classical simulation. We also considered the current quantum computing landscape shaped by the NISQ era and the field's movement from isolated circuit demonstrations toward hybrid quantum-classical computation. This included a review of quantum hardware providers, their technology roadmaps, the systems currently available to the public, their pricing models, and the limitations of cloud access. Finally, we reflected on the importance of remaining grounded in scientific evidence, staying curious, and asking the right questions.

Quantum computing contains many open problems across both its hardware and theoretical sides. Hardware challenges include quantum error correction, system scalability, noise reduction, control, and the large physical resources required for fault-tolerant computation. On the theoretical side, important questions concern when and why quantum algorithms may perform more efficiently than classical methods and why some quantum computations are difficult to simulate classically. Some partial results are known. For example, multipartite entanglement that grows with input size is necessary for exponential speedup in pure-state quantum algorithms, while Clifford circuits can contain entanglement and still be efficiently simulated classically. Each area in quantum computing also contains more focused questions that can be studied further. In quantum machine learning, for example, active research topics include barren plateaus, the expressive capacity of parameterized quantum models, and efficient methods for encoding classical data into quantum states. These engineering, theoretical, and applied questions leave room for further research and help keep quantum computing an active and evolving field.

1. **Early quantum computing, before the NISQ era.** Much of the early research focused on understanding what quantum algorithms are, identifying the problems for which they may outperform classical methods, and developing techniques that could be reused across different algorithms. This period produced foundational algorithms and methods such as phase estimation, amplitude amplification, and quantum Fourier transforms, together with small laboratory demonstrations. The primary goals were to lay the theoretical foundation and answer fundamental questions: What constitutes a quantum algorithm? Which problems are they suited for? Which problems offer no quantum advantage?

2. **The NISQ era.** The NISQ era is characterized by modest qubit counts, physical noise, limited coherence, and restrictions on circuit depth. The focus has shifted toward hybrid quantum-classical computation, identifying problems that fit within current hardware constraints, and improving device performance. Quantum hardware has also become commercially accessible through cloud services. Having established basic algorithmic capabilities, the field pursues proof-of-concept demonstrations at the edge of classical simulation, as well as small-scale quantum utility that pushes beyond the practical limits of classical simulation. Here, we try to answer questions like: How can we mitigate errors effectively without the overhead of full fault tolerance? How do we best partition computational tasks between quantum and classical processors to minimize noise-induced errors and maximize overall efficiency? Can we find practical problems where noisy quantum processors provide meaningful quantum utility?

3. **The early fault-tolerant quantum computing era.** In this stage, quantum error correction is expected to support longer and more reliable computations using logical qubits. Larger circuits could permit more substantial demonstrations of algorithms that depend on phase estimation, Hamiltonian simulation, and other deep quantum subroutines. Experimental implementations of Shor's algorithm may gradually increase in size. While early demonstrations of quantum advantage may occur, they will likely remain sporadic and confined to highly specialized, constrained problems. Based on the roadmaps of quantum hardware providers, early fault-tolerant systems are forecasted to emerge by the end of the 2020s or the beginning of the 2030s. During this era, research will focus on investigating questions such as: How do we optimize control systems and classical decoders to handle error correction in real time as circuits grow? What is a viable physical-to-logical qubit overhead required to run a useful fault-tolerant algorithm? Can we identify specialized problems in chemistry or materials science where early, unscaled fault-tolerant quantum computers might achieve a verifiable quantum advantage? 

4. **Long-term fault-tolerant quantum computing and beyond.** This stage would involve substantially larger numbers of physical and logical qubits, lower logical error rates, and enough fault-tolerant operations to run more demanding algorithms. It remains unclear which hardware platforms will scale most effectively or how current engineering challenges will be overcome. While long-term predictions are speculative, this era will likely see quantum computing expand into wider commercial sectors with tangible business value, moving beyond highly specialized physics or chemistry simulations. Progress toward factoring cryptographically relevant integers using Shor's algorithm would be one indication of reaching a much more advanced fault-tolerant regime. Quantum computers at this scale will likely rely heavily on hybrid quantum-classical integration. We may also see smaller, on-premise quantum devices that enable larger-scale quantum communication networks. While it is reasonable to speculate that some of these directions may begin to materialize around the mid-to-late 2030s, changes in hardware, error correction, algorithms, and classical computing may substantially alter these timelines. During this phase, research and development would likely shift toward studying questions such as: How will full-stack software architectures adapt to optimize workloads across global networks of classical supercomputers and interconnected quantum processors? What standards are needed to make algorithms, software, and workloads portable across hardware platforms? Which scientific and commercial problems justify the cost of fault-tolerant quantum computation? Can quantum hardware become smaller and sufficiently reliable for secure on-premises deployment?

    Over a still longer horizon, quantum processors could become more deeply integrated with classical computing systems, technological infrastructure, and communication networks. Secure quantum networks, large-scale distributed quantum computation, and closer integration between CPUs, GPUs, and QPUs may become possible. These possibilities are grounded in active research directions, but their eventual form and timeline are too uncertain to predict reliably.

All in all, quantum computing is an evolving and growing field with some answered questions and many more open questions and challenges. Usually, the more questions you study, the more questions arise. Improving and understanding quantum technologies requires discipline, perseverance, grit, and the will to learn, as well as curiosity, just for the chance to understand our mysterious universe a little better, advance humanity, and maybe leave a mark that you were here.

Be curious. Be bold. Take risks. Live life.
